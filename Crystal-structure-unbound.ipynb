```python
import pandas as pd
import numpy as np
from numpy.linalg import norm
import numpy.linalg
import MDAnalysis
import MDAnalysis.lib.distances
from scipy.spatial import distance
from numpy import *
from MDAnalysis.analysis import distances
from MDAnalysis.analysis.hydrogenbonds import HydrogenBondAnalysis
from MDAnalysis.analysis.dihedrals import Dihedral
from MDAnalysis.analysis.dihedrals import *
```

```python
HP1_1v49 = MDAnalysis.Universe('POPC-PV/Crystal-structure-PV/LC3B/unbound/1v49.pdb') #load PDB
residue_selection=[19,23,32,34,51,53,108] #Residues that need to be in the matrix

#Take COG of side chains

SC_COG_unb=[]
for i in residue_selection:
    sel1 = HP1_1v49.select_atoms( 'resid %i and (not (backbone or name HA or name HN) )'%i).center_of_geometry()
    SC_COG_unb.append(sel1)
SC_COG_unb=np.asarray(SC_COG_unb)

#Compute distances

ts_dist_1v49 = distance.cdist(SC_COG_unb, SC_COG_unb, 'euclidean')
ts_dist_1v49=np.tril(ts_dist_1v49,k=-1)
ts_dist_1v49
```

```python
import seaborn as sns

sns.heatmap(ts_dist_1v49,vmax=15,vmin=0,center=6,cmap='Blues',annot=True)
plt.title('1v49 HP1 Distance Matrix ')
plt.savefig('HP1_1v49.png',dpi=650,bbox_inches='tight')
```

```python
import pandas as pd
import numpy
import numpy.linalg
import MDAnalysis
import MDAnalysis.lib.distances
from scipy.spatial import distance
from numpy import *
from MDAnalysis.analysis import distances
from MDAnalysis.analysis.hydrogenbonds import HydrogenBondAnalysis
from MDAnalysis.analysis.dihedrals import Dihedral
from MDAnalysis.analysis.dihedrals import *
import holoviews as hv
from holoviews import opts, Cycle
from bokeh.io import export_svgs
hv.extension('bokeh')

import warnings
warnings.filterwarnings('ignore')
```

```python
u_1v49 = MDAnalysis.Universe("POPC-PV/Crystal-structure-PV/LC3B/unbound/1v49.pdb")

# r_unb = u_1v49.select_atoms("resid 19 or resid 23 or resid 32 or resid 34 or resid 51 or resid 53 or resid 108")
r_unb = u_1v49.select_atoms("resid 1-119")
J_UNB = Janin(r_unb).run()

z=len(J_UNB.ag4.residues)
chi1_1v49_df = pd.DataFrame()
for i in range (0,z-1):
    frames=[]
    ch1_unbound_ls=[]
#     cols_unbound = ['Frame', 'Chi1']
    
    for k,frame in zip(J_UNB.angles,J_UNB.frames):
        ch1_unbound=k[i:i+1,0][0]
        resn=J_UNB.ag4.resnames[i]
        resi=J_UNB.ag4.resids[i]
        ch1_unbound_ls.append(ch1_unbound)
        frames.append(frame)
        label_name='%d'%(resi)
    chi1_1v49_df['%s'%label_name]=ch1_unbound_ls
```

```python
chi1_1v49_df
```

```python
new_df=chi1_1v49_df[['19','23','34','51','53','108']]
new_df
```

```python
new_df=new_df.transpose()
new_df
```

```python
new_df.index
```

```python
plt.bar(new_df.index,new_df[0])
plt.ylim(0,360)
plt.xlabel('HP1 Residues')
plt.ylabel('Chi1')
plt.savefig('1v49_chi1.png',dpi=650,bbox_inches='tight')
```
