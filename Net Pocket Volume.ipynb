```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Net_Volume=pd.read_csv("Net_Volume_LC3B_new.csv")
Net_Volume
```

```python
start=0
time=[]
for i in range(1,1001):
    
    start=start+0.001
    time.append(start)
time.insert(0,0)
time[-1] #to check the last value
len(time)
```

```python
Net_Volume['Time']=time #to create a new column named Time and append the values to the column
Net_Volume
```

```python
plt.figure(figsize=(40,30))

sns.kdeplot(data=Net_Volume['proLC3B-S1'],fill=True,alpha=0.5,label="CS LC3B",color='#FF7878',linewidth=8)
sns.kdeplot(data=Net_Volume['proLC3B-S2'],fill=True,alpha=0.5,color='#FF7878',linewidth=8)
sns.kdeplot(data=Net_Volume['proLC3B-S3'],fill=True,alpha=0.5,color='#FF7878',linewidth=8)

sns.kdeplot(data=Net_Volume['LC3B-S1'],fill=True,alpha=0.5,label="Apo MB LC3B",color='#5A91FF',linewidth=8)
sns.kdeplot(data=Net_Volume['LC3B-S2'],fill=True,alpha=0.5,color='#5A91FF',linewidth=8)
sns.kdeplot(data=Net_Volume['LC3B-S3'],fill=True,alpha=0.5,color='#5A91FF',linewidth=8)

sns.kdeplot(data=Net_Volume['LC3B-p62-S1'],fill=True,alpha=0.5,label="LIR MB LC3B",color='#7DA629',linewidth=8)
sns.kdeplot(data=Net_Volume['LC3B-p62-S2'],fill=True,alpha=0.5,color='#7DA629',linewidth=8)
sns.kdeplot(data=Net_Volume['LC3B-p62-S3'],fill=True,alpha=0.5,color='#7DA629',linewidth=8)

plt.xlabel("Net Pocket Volume (Ã…\u00B3)", size=140)
plt.ylabel("Distribution", size=140)
plt.rcParams.update({'font.size': 120})

label_size = 100
plt.rcParams['xtick.labelsize'] = label_size
plt.rcParams['ytick.labelsize'] = label_size

plt.rcParams['xtick.bottom'] = True
plt.rcParams['xtick.top'] = True
plt.rcParams['xtick.major.size'] = 30
plt.rcParams['xtick.major.width'] = 10
plt.rcParams['xtick.direction'] = 'in'

plt.rcParams['ytick.left'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['ytick.major.size'] = 30
plt.rcParams['ytick.major.width'] = 10
plt.rcParams['ytick.direction'] = 'in'

plt.rcParams["figure.autolayout"] = True
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams["axes.linewidth"] = 8

# plt.xlim([-5,250])
# plt.ylim([0, 0.10])


plt.legend(loc='upper right', prop={'size':90})

plt.savefig("LC3B-net-pocket-volume_distribution.png",dpi=650,bbox_inches='tight')
```
